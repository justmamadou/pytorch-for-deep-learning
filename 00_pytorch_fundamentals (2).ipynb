### Matrix multiplication
There are two ways of performing  multiplication in neurla networks and deep learning:
1. Element-wise multiplication
2. Matrix multiplication (dot product)

There are two main rules that performing matrix multiplication needs to satisty:
1. The **inner dimensions** must match


*   (3,2) @ (3,2) won't work
*   (2,3) @ (3,2) will work
*   (3,2) @ (2,3) will work

2. The resulting matrix has the shape of the **outer dimensions**
*   (2,3) @ (3,2) -> (2,2)
*   (3,2) @ (2,3) -> (3,3)


```python
# Element wise multiplication
print(tensor, "*", tensor)
print(f"Equals: {tensor*tensor}")
```

```python
# matrix multiplication by hand
multi = 0
for i in range(len(tensor)):
  multi += tensor[i]*tensor[i]
print(multi)
```

```python
# Matrice multplication
matrice1 = torch.tensor([[1,2,3],[4,5,6]])
matrice2 = torch.tensor([[7,8,9],[10,11,12],[13,14,15]])

print(matrice1)
print(matrice2)
print(f"Shape of matrice 1: {matrice1.shape}")
print(f"Shape of matrice 2: {matrice2.shape}")


torch.matmul(matrice1, matrice2)
```

### One of the most common errors in deep learning: shape errors

```python
# Shapes for matrix multiplication
tensor_A = torch.tensor([[1,2],
                         [3,4],
                         [5,6]])
tensor_B = torch.tensor([[7,10],
                         [8,11],
                         [9,12]])

torch.matmul(tensor_A, tensor_B)
```

To fix our tensoir shape issue, we can manipulate the shape of one of our tensors using a **transpose**

A **transpose** switches the axes or dimensions of a given tensor

```python
tensor_B, tensor_B.shape
```

```python
tensor_B.T, tensor_B.T.shape
```

```python
# Matrix multiplication operation works when tensor_B is transposed

print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.T.shape}")

torch.matmul(tensor_A, tensor_B.T)
```

### Reshapping, stacking, squeezing and unsqueezing tensors

* Reshaping - reshapes an input tensor to a defined shape
* View - return a view of an input tensor of certain shape but keep the memory same as the original tensor
* Stacking - combine multiple tensors on top of each others (vstack) or side by side (hstack)
* Squeeze - remove all `1` dimensions from a tensor
* Unsqueeze - add a `1` dimension to a target tensor
* Permute - Return a view of the input with the dimensions permuted (swapped) in a certain way

```python
import torch

x = torch.arange(1., 10.)
x, x.shape
```

```python
## Add an extra dimension
x_reshaped = x.reshape(1,9)
x_reshaped, x_reshaped.shape
```

```python
# Change the viex
z = x.view(1,9)
z, z.shape
```

```python
# Changing z changes x (because a view of tensor shares the same memory as the original)
z[:, 0] = 5
z, x
```

```python
# Stack tensors on top on each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked
```

```python
# torch.Squeeze() - remove all single dimension
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

# Remvove extra dimensions from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"New tensor: {x_squeezed}")
print(f"Nex shape: {x_squeezed.shape}")
```

```python
# torch.unsqueeze() - adds a single dimension to a target tensor at a specific domension

print(f"Previous target: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

# Remvove extra dimensions from x_reshaped
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"New tensor: {x_unsqueezed}")
print(f"Nex shape: {x_unsqueezed.shape}")
```

```python
# torch.permute - rearranges the dimensions of a target tensor in a specified order
x_original = torch.rand(size=(224,224,3)) # [height, width, colour_channels]

# Permuted the original tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2,0,1) # shifts axis 0->1, 1->2, 2->0

print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}") # [colour_channels, height, width]
```

### Reproducbility (trying to take random out of random)

In short how a neural network learns:
`start with random numbers -> tensor operations -> update random numbers to try and make them better representations of the data -> again -> agin -> agin...`

To reduce the randomness in neural networks and PyTorch comes the concept of a **random sedd**.
Essentially what the random seed does is "flavour" the randomness

```python
import torch
# Create two random tensors

random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
# Let's make some random but reproducbility tensors

import torch


RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3, 4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
